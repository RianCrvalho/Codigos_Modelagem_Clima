```python
import pandas as pd
import matplotlib.pyplot as plt

# Criando um DataFrame de exemplo
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
data = {'Data': dates, 'Valor': range(len(dates))}
df = pd.DataFrame(data)

# Definindo a coluna 'Data' como ndice
df.set_index('Data', inplace=True)

# Agrupando por ms e somando os valores
monthly_sum = df.groupby(pd.Grouper(freq='M')).sum()

# Convertendo o ndice de datetime para nomes de ms completo
monthly_sum.index = monthly_sum.index.strftime('%B')

# Criando o grfico
plt.figure(figsize=(10, 6))
monthly_sum['Valor'].plot(kind='bar', color='skyblue')
plt.title('Soma Mensal dos Valores')
plt.xlabel('Ms')
plt.ylabel('Soma dos Valores')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```
