```python
#Plotagem das Barras e da média da Produção Agricola
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import leastsq
from numpy import array
from scipy.interpolate import interp1d
from datetime import datetime
import pandas as pd
%config InlineBackend.figure_format='retina'

# Ler os dados do arquivo Excel
df = pd.read_excel("Media_Quadrienio_soja.xlsx")

# Definindo os valores correspondentes aos meses
quadrienios = ['2008-2011', '2012-2015', '2016-2019', '2020-2023']

# Criando um dicionário para mapear os meses para seus respectivos números
quadrienios_dict = {mes: i+1 for i, mes in enumerate(quadrienios)}

# Tamanho da Imagem
plt.figure(figsize=(12, 6))

# Determinando uma média dos dados
Media= df["Med"].mean()

#Determinando barras abaixo e acima da média
for i in range(4):
    if df["Med"][i]>= Media:
        plt.bar(i+1, df["Med"][i]/100, color="#48D1CC") ##808080 cor original é o "c"
        
    else:
        plt.bar(i+1, df["Med"][i]/100, color='red')
        
#Implementação para a descrição das barras        
plt.bar(1, df["Med"][0]/100, color="#48D1CC",label="Acima da Média") #Tomar cuidado
plt.bar(2, df["Med"][1]/100, color="red",label="Abaixo da Média") #Tomar cuidado

#Plotando a linha média
plt.plot(np.linspace(0.6,4.4,5),(Media/100,Media/100,Media/100,Media/100,Media/100), 'k--',label='Média', linewidth=4) 

#Configurações do Gráfico
plt.xlabel('Quadriênios entre 2008 - 2023') 
plt.ylabel('Produção em t x 100')
plt.xticks(range(1,5),quadrienios, rotation=0)  # Definindo os meses no eixo x
plt.title("Média da Produção de Soja em Toneladas por Hectare por Quadriênio")
plt.legend()
plt.grid(False)

#Salvando e mostrando o gráfico
plt.savefig('Quadri_Agro.png', bbox_inches='tight',dpi=300)
plt.show()
```

```python
#Plotagem das curvas das funções da Produção de soja
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import leastsq
from numpy import array
from scipy.interpolate import interp1d
from datetime import datetime
import pandas as pd
%config InlineBackend.figure_format='retina'

# Ler os dados do arquivo Excel
df = pd.read_excel("Media_mensal_Intervalo_todo_soja.xlsx")
df0 = pd.read_excel("Media_mensal_Quadrienio0_soja.xlsx")
df1 = pd.read_excel("Media_mensal_Quadrienio1_soja.xlsx")
df2 = pd.read_excel("Media_mensal_Quadrienio2_soja.xlsx")
df3 = pd.read_excel("Media_mensal_Quadrienio3_soja.xlsx")


# Definindo os valores correspondentes aos meses
meses = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho", "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"]

# Criando um dicionário para mapear os meses para seus respectivos números
meses_dict = {mes: i+1 for i, mes in enumerate(meses)}

#Dividindo os valores da Produção Agricola /100 
df0['Med'] = df0['Med'] / 100
df1['Med'] = df1['Med'] / 100
df2['Med'] = df2['Med'] / 100
df3['Med'] = df3['Med'] / 100
df["Med"]  = df["Med"] /100

# Criando os novos valores de x para uma maior suavização da curva
xnew = np.arange(1, 12, 0.005)

# Interpolação
f0 = interp1d(list(range(1,13)), df0["Med"], kind='cubic')
f1 = interp1d(list(range(1,13)), df1["Med"], kind='cubic')
f2 = interp1d(list(range(1,13)), df2["Med"], kind='cubic')
f3 = interp1d(list(range(1,13)), df3["Med"], kind='cubic')

# Tamanho Imagem
plt.figure(figsize=(12, 6))

#Curvas com suas respectivas labels
plt.plot(xnew, f0(xnew), 'r', label='2007-2010')
plt.plot(xnew, f1(xnew), 'b', label='2011-2014')
plt.plot(xnew, f2(xnew), 'k', label='2015-2018')
plt.plot(xnew, f3(xnew), 'c', label='2019-2022')

#Pontos no grafico
plt.plot(range(1,13), df0["Med"], 'ro')
plt.plot(range(1,13), df1["Med"], 'bo')
plt.plot(range(1,13), df2["Med"], 'ko')
plt.plot(range(1,13), df3["Med"], 'co')

#Media dos meses ao fundo do gráfico
plt.bar(range(1,13),df["Med"], color="#D3D3D3", label="Média Mensal Total")

#Configurações do Gráfico
plt.title(r'$Variação\hspace{0.5} da\hspace{0.5} Média\hspace{0.5} de \hspace{0.5}Produção\hspace{0.5} de \hspace{0.5}Soja \hspace{0.5}Mensalmente$')
plt.xlabel(r'$Meses$')
plt.ylabel(r'$Produção\hspace{0.5} em \hspace{0.5}t\hspace{0.1}/\hspace{0.1}ha\hspace{0.5} x\hspace{0.5}100$')
plt.xticks(range(1,13),meses, rotation=20)  # Definindo os meses no eixo x
plt.legend()
plt.grid(False)

#Determinando intervalo do eixo y
plt.ylim([23, 34.7])

#Salvando e mostrando o gráfico
plt.savefig('meses_agro.png', bbox_inches='tight', pad_inches=0.1)
plt.show()
```
