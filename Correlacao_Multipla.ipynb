```python
#Correlação de todos os meses
import pandas as pd
import statsmodels.api as sm

# Suponha que 'data' é o DataFrame com as colunas 'Temperatura', 'Precipitação', 'Produção_Agricola'
dfa = pd.read_excel("Dados_Agricolas/Dados_Agricolas.xlsx")
dft = pd.read_excel("Dados_TEMP_MM/dados_dual.xlsx")


# Definindo a variável dependente
Y = dfa['Med']/10

# Definindo as variáveis independentes (com uma constante adicionada)
X = dft[['TEMP', 'MM']]
X = sm.add_constant(X)

# Ajustando o modelo de regressão linear
model = sm.OLS(Y, X).fit()

# Resumo do modelo
model_summary = model.summary()

# Coeficientes de regressão e R^2
coefficients = model.params
R_squared = model.rsquared

model_summary, coefficients, R_squared
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Ler o arquivo Excel
dfa = pd.read_excel("Dados_Agricolas/Dados_Agricolas.xlsx")
dft = pd.read_excel("Dados_TEMP_MM/dados_dual.xlsx")

# Separar as variáveis independentes e dependentes
y = dfa['Med']

# Definindo as variáveis independentes (com uma constante adicionada)
X = dft[['TEMP', 'MM']]

# Adicionando uma coluna de 1s para o intercepto
X = sm.add_constant(X)

# Ajustando o modelo de regressão linear
modelo = sm.OLS(y, X)
resultado = modelo.fit()

# Imprimindo os resultados
print(resultado.summary())

# Gráfico de dispersão dos dados observados versus valores preditos
plt.scatter(y, resultado.predict(X))
plt.xlabel('Produção Agrícola Observada')
plt.ylabel('Produção Agrícola Predita')
plt.title('Gráfico de Dispersão: Observado vs. Predito')
plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Ler o arquivo Excel
dfa = pd.read_excel("Dados_Agricolas/Dados_Agricolas.xlsx")
dft = pd.read_excel("Dados_TEMP_MM/dados_dual.xlsx")

# Separar as variáveis independentes e dependentes
y = dfa['Med']

# Definindo as variáveis independentes (com uma constante adicionada)
X = dft[['TEMP', 'MM']]

# Adicionando uma coluna de 1s para o intercepto
X = sm.add_constant(X)

# Ajustando o modelo de regressão linear
modelo = sm.OLS(y, X)
resultado = modelo.fit()

# Imprimindo os resultados
print(resultado.summary(title='Modelo da Regressão Multipla'))

# Gráfico de dispersão dos dados observados versus valores preditos
plt.scatter(y, resultado.predict(X), label='Observado vs. Predito')
plt.xlabel('Produção Agrícola Observada')
plt.ylabel('Produção Agrícola Predita')
# Adicionando a linha de regressão
plt.plot(y, y, color='red', label='Reta de Regressão')
plt.legend()
plt.xlim([950, 3750])
plt.ylim([2450, 3350])
plt.title('Gráfico de Dispersão: Observado vs. Predito com Reta de Regressão')
plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.iolib.summary2 import summary_col
%config InlineBackend.figure_format='retina'

# Ler os arquivos Excel
dfa = pd.read_excel("Dados_Agricolas/Dados_Agricolas.xlsx")
dft = pd.read_excel("Dados_TEMP_MM/dados_dual.xlsx")

# Separar a variável dependente
y = dfa['Med']

# Definir as variáveis independentes (com uma constante adicionada)
X = dft[['TEMP', 'MM']]
X = sm.add_constant(X)

# Ajustar o modelo de regressão linear
modelo = sm.OLS(y, X)
resultado = modelo.fit()

# Imprimir os resultados no console
print(resultado.summary(title='Modelo da Regressão Multipla'))

# Gráfico de dispersão dos dados observados versus valores preditos
plt.scatter(y, resultado.predict(X), label='Observado vs. Predito')
plt.xlabel('Produção Agrícola Observada')
plt.ylabel('Produção Agrícola Predita')

# Adicionar a linha de regressão
plt.plot(y, y, color='red', label='Reta de Regressão')
plt.legend()
plt.xlim([950, 3650])
plt.ylim([2200, 3500])
plt.title('Gráfico de Dispersão: Observado vs. Predito com Reta de Regressão')

# Salvar o gráfico de dispersão como uma imagem
plt.savefig('grafico_dispersao.png')
plt.show()

# Salvar o resumo da regressão em uma imagem
# Primeiro, vamos salvar o resumo em um arquivo de texto
with open('resumo_regressao.txt', 'w') as f:
    f.write(resultado.summary(title='Modelo da Regressão Multipla').as_text())

# Carregar o resumo salvo e plotá-lo como uma imagem
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO
import textwrap

# Lê o resumo do arquivo de texto
with open('resumo_regressao.txt', 'r') as f:
    resumo_texto = f.read()

# Quebra o texto em linhas para evitar que fique muito largo
wrapped_text = textwrap.fill(resumo_texto, width=100)

# Cria uma figura para a imagem do resumo
fig, ax = plt.subplots(figsize=(10, 6))
ax.text(0.01, 0.99, wrapped_text, transform=ax.transAxes, fontsize=10, verticalalignment='top')

# Remove os eixos
ax.axis('off')

# Salva a imagem do resumo
plt.savefig('resumo_regressao.png', bbox_inches='tight')
plt.show()
```
